Importing Libraries

```python
import spacy  #removing stopwords
from string import punctuation #removing punctuations
```

Reading the file 

```python
with open("./spaceX_DP.txt") as f: #opening text file 
    reviews = f.read() #reading the text file
```

```python
reviews
```

```python
en = spacy.load('en_core_web_sm')  #loading stopwords
stopwords = en.Defaults.stop_words
```

```python
lst=[]
for token in reviews.split():
    if token.lower() not in stopwords:
        lst.append(token)

print('Original Text') 
print("-------------")       
print(reviews,'\n\n')

print('Text after removing stop words')
print('------------------------------')
reviews_after_stopwords= (' '.join(lst)) #joing all the individual strings
print(reviews_after_stopwords)
```

Removing Punctuations

```python
reviews_after_punctuation = reviews_after_stopwords.translate(str.maketrans('','',punctuation))
```

```python
print('Text after removing punctuations')
print('---------------------------------')
print(reviews_after_punctuation)
```
